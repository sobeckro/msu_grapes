```python
import cv2
import numpy as np
import torch
from ultralytics import YOLO

# ---------------------------
# 1. Load models
# ---------------------------
det_model = YOLO("/content/drive/MyDrive/best.onnx", task="detect")

# SAM2 predictor must already be loaded:
# predictor = SAM2ImagePredictor(model)

# ---------------------------
# 2. Open video
# ---------------------------
cap = cv2.VideoCapture("row3_fixed.mp4")

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# ---------------------------
# 3. Setup output video
# ---------------------------
out = cv2.VideoWriter(
    "segmented_output_t7.mp4",
    cv2.VideoWriter_fourcc(*"mp4v"),
    fps,
    (width, height)
)

# ---------------------------
# 4. Initialize tracking + cache
# ---------------------------
frame_count = 0
mask_cache = {}  # store masks per tracked object

# ---------------------------
# 5. Process video frame-by-frame
# ---------------------------
while cap.isOpened():

    # Read next frame
    ret, frame = cap.read()
    if not ret:
        break

    frame_count += 1
    print("Processing frame:", frame_count)

    # ---------------------------
    # 6. YOLO detection + tracking
    # ---------------------------
    results = det_model.track(frame, persist=True)[0]

    if results.boxes is None or results.boxes.id is None:
        out.write(frame)
        continue

    boxes = results.boxes.xyxy.cpu().numpy()
    ids = results.boxes.id.cpu().numpy().astype(int)

    # ---------------------------
    # 7. Prepare overlay + SAM
    # ---------------------------
    overlay = frame.copy()
    alpha = 0.5

    predictor.set_image(frame)

    # ---------------------------
    # 8. Process each tracked object
    # ---------------------------
    for box, track_id in zip(boxes, ids):

        # If we've already segmented this object → reuse mask
        if track_id in mask_cache:
            mask = mask_cache[track_id]

        else:
            # ---------------------------
            # 9. Run SAM segmentation
            # ---------------------------
            masks, scores, logits = predictor.predict(
                box=np.array([box]),
                multimask_output=False
            )

            mask = masks[0].squeeze() > 0

            # Store mask for reuse
            mask_cache[track_id] = mask

        # ---------------------------
        # 10. Overlay segmentation mask
        # ---------------------------
        color = np.random.randint(0, 255, (3,), dtype=int)

        overlay[mask] = (
            overlay[mask] * (1 - alpha) + color * alpha
        ).astype(np.uint8)

    # ---------------------------
    # 11. Write frame to output video
    # ---------------------------
    out.write(overlay)

# ---------------------------
# 12. Release resources
# ---------------------------
cap.release()
out.release()
```

```python
import cv2
import numpy as np
import torch
from ultralytics import YOLO

# YOLO detection model
det_model = YOLO("/content/drive/MyDrive/best.onnx", task="detect")

cap = cv2.VideoCapture("row3_fixed.mp4")

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

out = cv2.VideoWriter(
    "segmented_output3.mp4",
    cv2.VideoWriter_fourcc(*"mp4v"),
    fps,
    (width, height)
)

frame_count = 0

# ----------------------------
# TRACKING STATE
# ----------------------------
prev_boxes = []
next_id = 0
IOU_THRESHOLD = 0.7
CENTROID_THRESHOLD = 30  # pixels
object_ids = {}  # maps box index → ID


def compute_iou(boxA, boxB):
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])

    interArea = max(0, xB - xA) * max(0, yB - yA)
    boxAArea = (boxA[2] - boxA[0]) * (boxA[3] - boxA[1])
    boxBArea = (boxB[2] - boxB[0]) * (boxB[3] - boxB[1])

    return interArea / float(boxAArea + boxBArea - interArea + 1e-6)


def centroid(box):
    x1, y1, x2, y2 = box
    return ((x1 + x2) / 2, (y1 + y2) / 2)


while cap.isOpened():

    ret, frame = cap.read()
    if not ret:
        break

    frame_count += 1
    print("Processing frame:", frame_count)

    # YOLO detection
    results = det_model(frame)[0]
    boxes = results.boxes.xyxy.cpu().numpy()

    overlay = frame.copy()
    alpha = 0.5

    predictor.set_image(frame)

    # ----------------------------
    # ASSIGN TRACK IDs USING IOU
    # ----------------------------
    current_ids = {}
    for i, box in enumerate(boxes):
        best_iou = 0
        best_id = None

        for j, prev_box in enumerate(prev_boxes):
            iou = compute_iou(box, prev_box)
            cx1, cy1 = centroid(box)
            cx2, cy2 = centroid(prev_box)
            shift = np.hypot(cx1 - cx2, cy1 - cy2)

            if iou > IOU_THRESHOLD and shift < CENTROID_THRESHOLD:
                if iou > best_iou:
                    best_iou = iou
                    best_id = object_ids.get(j)

        if best_id is None:
            best_id = next_id
            next_id += 1

        current_ids[i] = best_id

    # update tracking memory
    object_ids = current_ids
    prev_boxes = boxes.copy()

    # ----------------------------
    # SAM SEGMENTATION
    # ----------------------------
    if len(boxes) > 0:

        masks, scores, logits = predictor.predict(
            box=boxes,
            multimask_output=False
        )

        for (mask, box), obj_id in zip(zip(masks, boxes), current_ids.values()):

            mask = mask.squeeze() > 0
            color = np.random.randint(0,255,(3,),dtype=int)

            overlay[mask] = (
                overlay[mask] * (1 - alpha) + color * alpha
            ).astype(np.uint8)

            x1,y1,x2,y2 = box.astype(int)
            cv2.rectangle(overlay,(x1,y1),(x2,y2),(0,255,0),2)

            # draw ID
            cv2.putText(overlay, f"ID {obj_id}", (x1, y1-10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,255,0), 2)

    out.write(overlay)

cap.release()
out.release()
```

```python
import cv2
import numpy as np
import torch
from ultralytics import YOLO

# ---------------------------
# 1. Load YOLO model
# ---------------------------
det_model = YOLO("/content/drive/MyDrive/best.onnx", task="detect")

# SAM2 predictor assumed already loaded:
# predictor = SAM2ImagePredictor(model)

# ---------------------------
# 2. Open video
# ---------------------------
cap = cv2.VideoCapture("row1_fixed_longer.mp4")

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

out = cv2.VideoWriter(
    "segmented_output_FINAL.mp4",
    cv2.VideoWriter_fourcc(*"mp4v"),
    fps,
    (width, height)
)

# ---------------------------
# 3. Performance settings
# ---------------------------
FRAME_SKIP = 2          # process every 2nd frame
UPDATE_INTERVAL = 10    # refresh SAM every N frames

frame_count = 0
mask_cache = {}

# ---------------------------
# 4. Main loop
# ---------------------------
while cap.isOpened():

    ret, frame = cap.read()
    if not ret:
        break

    frame_count += 1

    # Skip frames for speed
    if frame_count % FRAME_SKIP != 0:
        continue

    print("Processing frame:", frame_count)

    # ---------------------------
    # 5. YOLO tracking
    # ---------------------------
    results = det_model.track(frame, persist=True)[0]

    if results.boxes is None or results.boxes.id is None:
        out.write(frame)
        continue

    boxes = results.boxes.xyxy.cpu().numpy()
    ids = results.boxes.id.cpu().numpy().astype(int)

    overlay = frame.copy()
    alpha = 0.5

    predictor.set_image(frame)

    # ---------------------------
    # 6. Process each object
    # ---------------------------
    for box, track_id in zip(boxes, ids):

        x1, y1, x2, y2 = box

        # 🎯 CENTER POINT (better than box for grapes)
        cx = int((x1 + x2) / 2)
        cy = int((y1 + y2) / 2)

        # ---------------------------
        # 7. Decide whether to update mask
        # ---------------------------
        if (
            track_id not in mask_cache or
            frame_count % UPDATE_INTERVAL == 0
        ):
            masks, scores, logits = predictor.predict(
                point_coords=np.array([[cx, cy]]),
                point_labels=np.array([1]),  # positive point
                multimask_output=False
            )

            mask = masks[0].squeeze() > 0
            mask_cache[track_id] = mask

        else:
            mask = mask_cache[track_id]

        # ---------------------------
        # 8. Apply segmentation
        # ---------------------------
        color = np.array([0, 255, 0])  # consistent green

        overlay[mask] = (
            overlay[mask] * (1 - alpha) + color * alpha
        ).astype(np.uint8)

    # ---------------------------
    # 9. Write output frame
    # ---------------------------
    out.write(overlay)

# ---------------------------
# 10. Cleanup
# ---------------------------
cap.release()
out.release()
```

```python
import cv2
import numpy as np
import torch
from ultralytics import YOLO

# ---------------------------
# Load YOLO
# ---------------------------
det_model = YOLO("/content/drive/MyDrive/best.onnx", task="detect")

cap = cv2.VideoCapture("row3_fixed.mp4")

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

out = cv2.VideoWriter(
    "segmented_tracked2_32026.mp4",
    cv2.VideoWriter_fourcc(*"mp4v"),
    fps,
    (width, height)
)

frame_count = 0

# ----------------------------
# TRACKING STATE
# ----------------------------
tracked_objects = {}   # id → {centroid, missing}
next_id = 0

MAX_DISTANCE = 100   # how far an object can move
MAX_MISSING = 10     # how long to keep "lost" objects


def centroid(box):
    x1, y1, x2, y2 = box
    return ((x1 + x2) / 2, (y1 + y2) / 2)


while cap.isOpened():

    ret, frame = cap.read()
    if not ret:
        break

    frame_count += 1
    print("Processing frame:", frame_count)

    # YOLO detection
    results = det_model(frame)[0]
    boxes = results.boxes.xyxy.cpu().numpy()

    overlay = frame.copy()
    alpha = 0.5

    # ----------------------------
    # TRACKING (FIXED VERSION)
    # ----------------------------
    current_ids = {}
    used_ids = set()

    for i, box in enumerate(boxes):

        cx, cy = centroid(box)

        best_id = None
        best_dist = float("inf")

        # match to previous objects
        for obj_id, data in tracked_objects.items():
            prev_cx, prev_cy = data["centroid"]

            dist = np.hypot(cx - prev_cx, cy - prev_cy)

            if dist < best_dist and dist < MAX_DISTANCE and obj_id not in used_ids:
                best_dist = dist
                best_id = obj_id

        # assign new ID if no match
        if best_id is None:
            best_id = next_id
            next_id += 1

        current_ids[i] = best_id
        used_ids.add(best_id)

        # update tracked object
        tracked_objects[best_id] = {
            "centroid": (cx, cy),
            "missing": 0
        }

    # remove lost objects
    for obj_id in list(tracked_objects.keys()):
        if obj_id not in used_ids:
            tracked_objects[obj_id]["missing"] += 1

            if tracked_objects[obj_id]["missing"] > MAX_MISSING:
                del tracked_objects[obj_id]

    # ----------------------------
    # SAM SEGMENTATION
    # ----------------------------
    predictor.set_image(frame)

    if len(boxes) > 0:

        masks, scores, logits = predictor.predict(
            box=boxes,
            multimask_output=False
        )

        for (mask, box), obj_id in zip(zip(masks, boxes), current_ids.values()):

            mask = mask.squeeze() > 0

            # stable color per ID
            np.random.seed(obj_id)
            color = np.random.randint(0,255,(3,),dtype=int)

            overlay[mask] = (
                overlay[mask] * (1 - alpha) + color * alpha
            ).astype(np.uint8)

            # OPTIONAL: draw ID
            x1,y1,x2,y2 = box.astype(int)
            cv2.putText(overlay, f"ID {obj_id}", (x1, y1-10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255,255,255), 2)

    out.write(overlay)

cap.release()
out.release()
```
